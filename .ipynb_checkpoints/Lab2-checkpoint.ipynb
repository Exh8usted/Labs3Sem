Считывание файла

```python
import pandas
import os

file_path = "Cleaned_ships_data.csv"
ram_capacity = 2**20

if file_path.endswith('.csv') and os.path.isfile(file_path) and os.path.getsize(file_path) <= ram_capacity:
    ship_data = pandas.read_csv(file_path)
    if not (ship_data.empty or ship_data.isnull().values.any()):
        print(ship_data)
    else:
        print("БИТЫЙ ФАЙЛ")
else:
    print("НЕТ ТАКОГО ФАЙЛА")
```

2-D график зависимости грузоподъемности (dwt) от длины (length)

```python
import matplotlib.pyplot as plt

sorted_ship_data = ship_data.sort_values('length')

plt.figure(figsize = (20, 10))

plt.plot(sorted_ship_data.length, sorted_ship_data.dwt, '-o', color = 'pink')

plt.title('Зависимость грузоподъемности от длины')
plt.xlabel("Длина коробля")
plt.ylabel("Грузоподъемность")
plt.grid(True)
plt.xticks(range(270, 490, 5))

plt.show()



sorted_ship_data = ship_data.sort_values('width')

plt.figure(figsize = (15, 10))

plt.plot(sorted_ship_data.width, sorted_ship_data.dwt, '-o', color = 'pink')

plt.title('Зависимость грузоподъемности от ширины')
plt.xlabel("Ширина коробля")
plt.ylabel("Грузоподъемность")
plt.grid(True)
plt.xticks(range(35, 76, 2))

plt.show()
```

Распределение кораблей по виду (ship_name)

```python
import matplotlib.pyplot as plt

list_of_names = ship_data.ship_name
names_with_count = list_of_names.value_counts()

plt.figure(figsize = (15, 10))

plt.plot(names_with_count, '-o', color = 'pink')

plt.title('Распределение кораблей по виду')
plt.xlabel("Вид корабля")
plt.ylabel("Количество")
plt.grid(True)

plt.show()
```

Гистограмма распределения количества кораблей по годам постройки (built_year)

```python
import matplotlib.pyplot as plt

list_of_years = ship_data.built_year
years_with_count = list_of_years.value_counts()

plt.figure(figsize=(10, 10))

plt.bar(years_with_count.index, years_with_count.values, color = 'pink')

plt.title('Количество кораблей по годам постройки')
plt.xlabel("Года")
plt.ylabel("Количество кораблей")

plt.show()
```

2-D Гистограмма распределения количества кораблей по годам 
постройки (built_year) и максимальному весу груза (dwt)

```python
import matplotlib.pyplot as plt

list_of_years = ship_data.built_year
list_of_dwts = ship_data.dwt

plt.hist2d(list_of_years, list_of_dwts, bins = 20, cmap = 'magma')

plt.title('Количество кораблей по годам постройки и максимальному весу груза')
plt.xlabel('Годы')
plt.ylabel('Максимальный вес груза')

plt.colorbar()

plt.show()
```

Задание 2. Использованный датасет: https://www.kaggle.com/datasets/harishkumardatalab/data-science-salary-2021-to-202


Считывание файла

```python
import pandas
import os

file_path = "Data Science Salary 2021 to 2023.csv"
ram_capacity = 2**20

if file_path.endswith('.csv') and os.path.isfile(file_path) and os.path.getsize(file_path) <= ram_capacity:
    salary_data = pandas.read_csv(file_path)
    if not (salary_data.empty or salary_data.isnull().values.any()):
        print(salary_data)
    else:
        print("БИТЫЙ ФАЙЛ")
else:
    print("НЕТ ТАКОГО ФАЙЛА")
```

Гистограмма средней зарплаты в долларах (salary_in_usd) для каждого уровня опытности сотрудников (experience_level)

```python
import pandas
import matplotlib.pyplot as plt

mean_salary_by_experience = salary_data.groupby('experience_level')['salary_in_usd'].median()

plt.bar(mean_salary_by_experience.index, mean_salary_by_experience.values, color = 'pink')

plt.title('Средняя зарплата в зависимости от опытности сотрудника')
plt.xlabel('Уровень сотрудника')
plt.ylabel('Средняя зарплата в долларах')
plt.grid(True)

plt.show()


mean_salary_by_experience = salary_data.groupby('experience_level')['salary_in_usd'].mean()

plt.bar(mean_salary_by_experience.index, mean_salary_by_experience.values, color = 'pink')

plt.title('Средняя зарплата в зависимости от опытности сотрудника')
plt.xlabel('Уровень сотрудника')
plt.ylabel('Средняя зарплата в долларах')
plt.grid(True)

plt.show()
# Уровни сотрудника по возрастанию:
# EN (Entry-Level) - джун
# MI (Mid-Level) - мид
# EX (Experienced) - что-то между мидом и сеньором
# SE (Senior) - сеньор
# P.S. Забавно, что сеньоры получают меньше experienced
```

График распространенности различных вакансий (job_title)

```python
import matplotlib.pyplot as plt

list_of_titles = salary_data.job_title
titles_with_count = list_of_titles.value_counts()
mask = list_of_titles.isin(titles_with_count.index[titles_with_count < 40])
salary_data.loc[mask, 'job_title'] = 'Others'
titles_with_count = salary_data['job_title'].value_counts()

plt.figure(figsize = (20, 10))

plt.plot(titles_with_count, '-o', color = 'pink')

plt.title('Распространенность вакансий')
plt.xlabel('Вакансии')
plt.ylabel('Количество работников')
plt.grid(True)

plt.show()
```

2-D Гистограмма распределения количества работников по годам (work_year) и зарплате в долларах (salary_in_usd)

```python
import matplotlib.pyplot as plt

list_of_work_years = salary_data.work_year
list_of_salary_in_usd = salary_data.salary_in_usd

plt.figure(figsize = (10,4))

plt.hist2d(list_of_work_years, list_of_salary_in_usd, bins = 20, cmap = 'magma')

plt.title('Количество сотрудников по годам и зарплате')
plt.xlabel('Годы')
plt.ylabel('Зарплата в долларах')

plt.colorbar()

plt.show()
```
