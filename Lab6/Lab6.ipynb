```python
import subprocess

c_sharp_code_file_path = r"./Demographic.Exec/Demographic.Exec.csproj"
initial_age_file_path = "./InitialAge.csv"
death_rules_file_path = "./DeathRules.csv"
ages_result_file_path = "./ages_result.csv"
population_result_file_path = "./population_result.csv"
start_year = "1970"
end_year = "2021"
initial_people_amount = "25"
subprocess.run(["dotnet", "run", "--project", c_sharp_code_file_path, initial_age_file_path, death_rules_file_path
              , ages_result_file_path, population_result_file_path, start_year, end_year, initial_people_amount])
```

```python
import pandas

ages_data = pandas.read_csv(ages_result_file_path)
population_data = pandas.read_csv(population_result_file_path)

print(ages_data, "\n\n", population_data)

# 0   11785   11908  23693
# 1   11363   11541  22904
# 2   11086   11317  22403
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

plt.plot([i for i in range(int(start_year), int(end_year))], population_data["All"], "-o")

plt.title("График общей популяции населения по годам")
plt.xlabel("Годы")
plt.ylabel("Популяция (тыс. чел)")
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

plt.plot([i for i in range(int(start_year), int(end_year))], population_data["Male"], "-o")

plt.title("График популяции мужчин по годам")
plt.xlabel("Годы")
plt.ylabel("Популяция (тыс. чел)")
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

plt.plot([i for i in range(int(start_year), int(end_year))], population_data["Female"], "-o")

plt.title("График популяции женщин по годам")
plt.xlabel("Годы")
plt.ylabel("Популяция (тыс. чел)")
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

formatted_ages_data = ages_data.drop('Gender', axis=1).drop(1)

plt.figure(figsize=(10, 10))

plt.bar(formatted_ages_data.columns, formatted_ages_data.iloc[0], color = 'pink')

plt.title("Распределение возрастов мужского населения")
plt.xlabel("Возраст")
plt.ylabel("Популяция (тыс. чел)")
plt.grid(True)

plt.show()
```
